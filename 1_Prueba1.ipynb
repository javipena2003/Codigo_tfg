A lo largo de este documento se muestran ejemplos del uso de la biblioteca de Bloquade: Establecer geometrías, Declarar waveforms, creación del hamiltoniano a partir de simular un ejemplo

```python
# =============================================================
# Chapter I - Basics of Bloqade (QuEra Education Tutorial)
# =============================================================
from bloqade.analog import rydberg_h, piecewise_linear, piecewise_constant, waveform, cast
from bloqade.analog.atom_arrangement import ListOfLocations, Lieb, Square, Chain, Honeycomb, Kagome, Triangular, Rectangular
from bokeh.io import output_notebook # to plot "show()" on the notebook, without opening a browser
import os
import matplotlib.pyplot as plt
import numpy as np
output_notebook()
```

# 1. Establecer geometrías iniciales

```python
my_register = ListOfLocations([(0.0, 0.0), (0.0, 5.0), (0.0, 9.0), (5.0, 2.0), (6.0, 7.0), (9.0, 10.0)])
my_register.show()

Square(4,3,lattice_spacing=5.2).show()

rng = np.random.default_rng(1234)
Honeycomb(2,2, lattice_spacing=4.5).apply_defect_density(0.1, rng=rng).remove_vacant_sites().show() # Chain, Lieb, Honeycomb, Kagome, Triangular, Rectangular
```

# 2. Declaring waveforms

```python
durations = [0.4,3.2,0.4]
values_MHz = [0.0,2.5,2.5,0.0]
values_radsec = [i * 2*np.pi for i in values_MHz]
waveform1 = piecewise_linear(durations, values_radsec)
waveform1.show()


durations = [2.0,2.0]
values = [0.0,np.pi]
waveform2 = piecewise_constant(durations, values)
waveform2.show()
```

# 3. Creación del Hamiltoniano
Establecemos todas las variables desde cero

```python
#geometry
rng = np.random.default_rng(1234)
#atom_pos=Square(4, lattice_spacing=4.5).apply_defect_density(0.3, rng=rng).remove_vacant_sites()
atom_pos = Honeycomb(2,2, lattice_spacing=4.5).apply_defect_density(0.2, rng=rng).remove_vacant_sites() # Chain, Lieb, Honeycomb, Kagome, Triangular, Rectangular

#dynamics
durations = [0.15,3.7,0.15]
delta_MHz=[-13.0,-13.0,11.0,11.0]
omega_MHz= [0.0,2.5,2.5,0.0]

Delta = piecewise_linear(durations,[x*2*np.pi for x in delta_MHz])
Omega = piecewise_linear(durations,[x*2*np.pi for x in omega_MHz])

#create Hamiltonian
program = rydberg_h(atom_pos, detuning= Delta, amplitude=Omega, phase=None)

program.parse_register().show()
program.parse_sequence().show()
```

```python
output = program.bloqade.python().run(shots=10, interaction_picture=True)
output.report().show()
```
